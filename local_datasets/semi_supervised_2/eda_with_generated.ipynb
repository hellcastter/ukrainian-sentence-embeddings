```python
import json

PATH = "merged_collected_and_generated_mpnet.json"
```

```python
with open(PATH, 'r') as f:
    data = json.load(f)
```

```python
# count number of sentences
counter = 0

for lemma, meanings in data.items():
    for meaning, meaning_data in meanings.items():
        counter += len(meaning_data["sentences"])
        
counter
```
